```python
import zipfile
import tensorflow as tf
from tensorflow.keras import layers, models, applications
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import io
import os
import random

# ============= Enhanced Constants =============
IMG_HEIGHT = 224
IMG_WIDTH = 224
NUM_LAYERS = 13
NUM_PATHOLOGIES = 4
BATCH_SIZE = 1    # Keep low due to model complexity
EPOCHS = 15       # Reduced for initial testing
ZIP_PATH = r'F:\gpu\OCTAVE.zip'

# ============= Advanced Loss Functions =============
def focal_loss(gamma=2.0, alpha=0.25):
    """Focal Loss for addressing class imbalance"""
    def focal_loss_fixed(y_true, y_pred):
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)
        
        cross_entropy = -y_true * tf.math.log(y_pred)
        weight = alpha * tf.math.pow((1 - y_pred), gamma)
        loss = weight * cross_entropy
        return tf.reduce_mean(loss)
    return focal_loss_fixed

def dice_loss(y_true, y_pred, smooth=1e-6):
    """Dice Loss for better segmentation"""
    y_true_f = tf.reshape(y_true, [-1])
    y_pred_f = tf.reshape(y_pred, [-1])
    intersection = tf.reduce_sum(y_true_f * y_pred_f)
    return 1 - (2. * intersection + smooth) / (tf.reduce_sum(y_true_f) + tf.reduce_sum(y_pred_f) + smooth)

# ============= Advanced Metrics =============
class DiceMetric(tf.keras.metrics.Metric):
    def __init__(self, name='dice_coef', **kwargs):
        super().__init__(name=name, **kwargs)
        self.total = self.add_weight(name='total', initializer='zeros')
        self.count = self.add_weight(name='count', initializer='zeros')

    def update_state(self, y_true, y_pred, sample_weight=None):
        epsilon = 1e-6
        y_true_f = tf.keras.backend.flatten(y_true)
        y_pred_f = tf.keras.backend.flatten(y_pred)
        intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
        dice = (2. * intersection + epsilon) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + epsilon)
        self.total.assign_add(dice)
        self.count.assign_add(1.0)

    def result(self):
        return self.total / self.count

    def reset_state(self):
        self.total.assign(0.0)
        self.count.assign(0.0)

def pixel_accuracy(y_true, y_pred):
    """Enhanced pixel accuracy calculation"""
    y_true_class = tf.argmax(y_true, axis=-1)
    y_pred_class = tf.argmax(y_pred, axis=-1)
    return tf.reduce_mean(tf.cast(tf.equal(y_true_class, y_pred_class), tf.float32))

# ============= Enhanced Dataset Loader =============
class OCTAVEDatasetFromZip:
    def __init__(self, zip_path):
        self.zip_path = zip_path
        self.train_images = []
        self.train_labels = []
        self.val_images = []
        self.val_labels = []
        
        with zipfile.ZipFile(zip_path, 'r') as z:
            all_files = z.namelist()
            self.train_images = sorted([f for f in all_files 
                                      if 'Dataset001_OCTAVE/imagesTr/' in f and f.endswith('.tif')])
            self.train_labels = sorted([f for f in all_files 
                                      if 'Dataset001_OCTAVE/labelsTr/' in f and f.endswith('.tif')])
            self.val_images = sorted([f for f in all_files 
                                    if 'Dataset001_OCTAVE/imagesTs/' in f and f.endswith('.tif')])
            self.val_labels = sorted([f for f in all_files 
                                    if 'Dataset001_OCTAVE/labelsTs/' in f and f.endswith('.tif')])

        if len(self.val_images) == 0:
            combined = list(zip(self.train_images, self.train_labels))
            random.shuffle(combined)
            val_size = int(len(combined) * 0.2)
            
            val_data = combined[:val_size]
            train_data = combined[val_size:]
            
            self.train_images, self.train_labels = zip(*train_data) if train_data else ([], [])
            self.val_images, self.val_labels = zip(*val_data) if val_data else ([], [])
            
            self.train_images = list(self.train_images)
            self.train_labels = list(self.train_labels)
            self.val_images = list(self.val_images)
            self.val_labels = list(self.val_labels)
            
        print(f"Train: {len(self.train_images)}, Val: {len(self.val_images)}")
    
    def load_image_from_zip(self, zip_file, filepath):
        """Load and process single-channel grayscale images"""
        try:
            with zip_file.open(filepath) as file:
                img_data = file.read()
                
                try:
                    img = Image.open(io.BytesIO(img_data))
                    img = img.convert('L')  # Convert to grayscale
                    img = img.resize((IMG_WIDTH, IMG_HEIGHT))
                    img = np.array(img, dtype=np.float32) / 255.0
                    
                    # Keep as single channel (224, 224, 1)
                    if img.ndim == 2:
                        img = np.expand_dims(img, axis=-1)
                    
                    if np.max(img) == 0:
                        return None
                    return img
                    
                except:
                    try:
                        import tifffile
                        img = tifffile.imread(io.BytesIO(img_data))
                        img = tf.image.resize(img, [IMG_HEIGHT, IMG_WIDTH], method='nearest').numpy()
                        
                        if img.ndim == 3:
                            img = np.mean(img, axis=2)  # Convert to grayscale
                        
                        if img.ndim == 2:
                            img = np.expand_dims(img, axis=-1)
                        
                        img = img.astype(np.float32)
                        if np.max(img) > 1.0:
                            img = img / 255.0
                            
                        if np.max(img) == 0:
                            return None
                        return img
                        
                    except:
                        return None
        except Exception as e:
            return None
    
    def validate_and_normalize_mask(self, mask):
        """Enhanced mask validation and normalization"""
        if mask is None:
            return None
            
        # Ensure single channel mask
        if mask.ndim == 3 and mask.shape[2] > 1:
            mask = mask[:, :, 0:1]  # Take first channel
        elif mask.ndim == 2:
            mask = np.expand_dims(mask, axis=-1)
        
        # Create synthetic multi-class labels
        layer_mask = np.zeros((mask.shape[0], mask.shape[1], NUM_LAYERS))
        pathology_mask = np.zeros((mask.shape[0], mask.shape[1], NUM_PATHOLOGIES))
        
        mask_values = mask.squeeze()
        
        # Create layer segmentation based on intensity thresholds
        for i in range(NUM_LAYERS):
            if i == 0:  # Background
                layer_mask[:, :, i] = (mask_values < 0.05).astype(np.float32)
            else:
                # Distribute layers across intensity ranges
                lower_threshold = (i - 1) / (NUM_LAYERS - 1) * 0.95 + 0.05
                upper_threshold = i / (NUM_LAYERS - 1) * 0.95 + 0.05
                layer_mask[:, :, i] = ((mask_values >= lower_threshold) & 
                                      (mask_values < upper_threshold)).astype(np.float32)
        
        # Create pathology masks
        pathology_mask[:, :, 0] = (mask_values > 0.7).astype(np.float32)  # Severe pathology
        pathology_mask[:, :, 1] = ((mask_values > 0.5) & (mask_values <= 0.7)).astype(np.float32)  # Moderate
        pathology_mask[:, :, 2] = ((mask_values > 0.3) & (mask_values <= 0.5)).astype(np.float32)  # Mild
        pathology_mask[:, :, 3] = ((mask_values > 0.1) & (mask_values <= 0.3)).astype(np.float32)  # Minimal
        
        # Normalize layer masks
        layer_sum = np.sum(layer_mask, axis=-1, keepdims=True)
        layer_sum = np.where(layer_sum == 0, 1, layer_sum)
        layer_mask = layer_mask / layer_sum
        
        # Clip pathology masks
        pathology_mask = np.clip(pathology_mask, 0, 1)
        
        normalized_mask = np.concatenate([layer_mask, pathology_mask], axis=-1)
        
        return normalized_mask
    
    def data_generator(self, split='train'):
        if split == 'train':
            images, labels = self.train_images, self.train_labels
        else:
            images, labels = self.val_images, self.val_labels
            
        successful_loads = 0
        
        with zipfile.ZipFile(self.zip_path, 'r') as zip_file:
            for img_path, label_path in zip(images, labels):
                image = self.load_image_from_zip(zip_file, img_path)
                mask = self.load_image_from_zip(zip_file, label_path)
                
                if image is not None and mask is not None:
                    mask = self.validate_and_normalize_mask(mask)
                    
                    if mask is not None:
                        successful_loads += 1
                        yield image, mask
        
        print(f"Successfully loaded {successful_loads} pairs for {split}")

# ============= Fixed Model Architecture =============
def conv_block(x, filters, kernel_size=3):
    """Enhanced convolution block"""
    x = layers.Conv2D(filters, kernel_size, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    
    x = layers.Conv2D(filters, kernel_size, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    
    return x

def encoder_block(x, filters):
    """Enhanced encoder block with downsampling"""
    conv = conv_block(x, filters)
    pool = layers.MaxPooling2D(pool_size=(2, 2))(conv)
    pool = layers.Dropout(0.1)(pool)
    return conv, pool

def decoder_block(x, skip, filters):
    """Enhanced decoder block with upsampling"""
    # Upsample
    up = layers.UpSampling2D(size=(2, 2), interpolation='bilinear')(x)
    
    # Ensure compatible dimensions
    if up.shape[1] != skip.shape[1] or up.shape[2] != skip.shape[2]:
        up = layers.Cropping2D(cropping=((0, max(0, up.shape[1] - skip.shape[1])), 
                                        (0, max(0, up.shape[2] - skip.shape[2]))))(up)
    
    # Concatenate
    merge = layers.Concatenate()([up, skip])
    
    # Apply convolution
    conv = conv_block(merge, filters)
    conv = layers.Dropout(0.1)(conv)
    
    return conv

def build_complete_lapm():
    """Build complete LAPM with U-Net architecture"""
    # Single-channel grayscale input
    inputs = layers.Input(shape=(IMG_HEIGHT, IMG_WIDTH, 1), name='Grayscale_Input')
    
    # Convert grayscale to 3-channel for better feature extraction
    rgb_inputs = layers.Concatenate(axis=-1, name='RGB_Converter')([inputs, inputs, inputs])
    
    # Encoder path
    conv1, pool1 = encoder_block(rgb_inputs, 64)      # 224 -> 112
    conv2, pool2 = encoder_block(pool1, 128)          # 112 -> 56
    conv3, pool3 = encoder_block(pool2, 256)          # 56 -> 28
    conv4, pool4 = encoder_block(pool3, 512)          # 28 -> 14
    
    # Bottleneck
    bottleneck = conv_block(pool4, 1024)              # 14
    bottleneck = layers.Dropout(0.2)(bottleneck)
    
    # Decoder path
    dec4 = decoder_block(bottleneck, conv4, 512)      # 14 -> 28
    dec3 = decoder_block(dec4, conv3, 256)            # 28 -> 56
    dec2 = decoder_block(dec3, conv2, 128)            # 56 -> 112
    dec1 = decoder_block(dec2, conv1, 64)             # 112 -> 224
    
    # Output heads
    # Layer segmentation head
    layer_features = conv_block(dec1, 64)
    layer_output = layers.Conv2D(NUM_LAYERS, 1, activation='softmax', name='layer_output')(layer_features)

    # Pathology segmentation head
    pathology_features = conv_block(dec1, 64)
    pathology_output = layers.Conv2D(NUM_PATHOLOGIES, 1, activation='sigmoid', name='pathology_output')(pathology_features)

    # Refined pathology with layer guidance
    layer_guidance = layers.Conv2D(32, 1, activation='relu')(layer_output)
    combined_features = layers.Concatenate()([pathology_features, layer_guidance])
    combined_features = conv_block(combined_features, 64)
    refined_pathology = layers.Conv2D(NUM_PATHOLOGIES, 1, activation='sigmoid', name='refined_pathology')(combined_features)

    model = models.Model(
        inputs=inputs,
        outputs=[layer_output, pathology_output, refined_pathology],
        name='Complete_LAPM'
    )
    
    return model

# ============= Enhanced Training Class =============
class CompleteLAPMTrainer(tf.keras.Model):
    def __init__(self, lapm_model):
        super().__init__()
        self.lapm = lapm_model
        
    def call(self, inputs, training=False):
        return self.lapm(inputs, training=training)
        
    def compile(self, optimizer, **kwargs):
        super().compile(**kwargs)
        self.optimizer = optimizer
        
        # Enhanced loss functions
        self.focal_loss = focal_loss(gamma=2.0, alpha=0.25)
        self.binary_crossentropy = tf.keras.losses.BinaryCrossentropy()
        
        # Comprehensive metrics
        self.loss_tracker = tf.keras.metrics.Mean(name="loss")
        self.layer_loss_tracker = tf.keras.metrics.Mean(name="layer_loss")
        self.pathology_loss_tracker = tf.keras.metrics.Mean(name="pathology_loss")
        self.refined_loss_tracker = tf.keras.metrics.Mean(name="refined_loss")
        
        # Advanced metrics
        self.layer_accuracy = tf.keras.metrics.Mean(name="layer_accuracy")
        self.layer_dice = DiceMetric(name="layer_dice")
        self.pathology_dice = DiceMetric(name="pathology_dice")
        self.refined_dice = DiceMetric(name="refined_dice")

    @property
    def metrics(self):
        return [
            self.loss_tracker, self.layer_loss_tracker, self.pathology_loss_tracker, self.refined_loss_tracker,
            self.layer_accuracy, self.layer_dice, self.pathology_dice, self.refined_dice
        ]

    def train_step(self, data):
        x, y = data
        
        layer_true = y[:, :, :, :NUM_LAYERS]
        patho_true = y[:, :, :, NUM_LAYERS:]
        
        with tf.GradientTape() as tape:
            layer_pred, patho_pred, refined_patho = self.lapm(x, training=True)
            
            # Enhanced loss computation
            layer_loss = tf.reduce_mean(self.focal_loss(layer_true, layer_pred))
            layer_loss += dice_loss(layer_true, layer_pred)
            
            patho_loss = tf.reduce_mean(self.binary_crossentropy(patho_true, patho_pred))
            patho_loss += dice_loss(patho_true, patho_pred)
            
            refined_loss = tf.reduce_mean(self.binary_crossentropy(patho_true, refined_patho))
            refined_loss += dice_loss(patho_true, refined_patho)
            
            total_loss = layer_loss + 2.0 * patho_loss + 1.5 * refined_loss
        
        gradients = tape.gradient(total_loss, self.lapm.trainable_variables)
        gradients = [tf.clip_by_norm(g, 1.0) for g in gradients]
        self.optimizer.apply_gradients(zip(gradients, self.lapm.trainable_variables))
        
        # Update metrics
        self.loss_tracker.update_state(total_loss)
        self.layer_loss_tracker.update_state(layer_loss)
        self.pathology_loss_tracker.update_state(patho_loss)
        self.refined_loss_tracker.update_state(refined_loss)
        
        self.layer_accuracy.update_state(pixel_accuracy(layer_true, layer_pred))
        self.layer_dice.update_state(layer_true, layer_pred)
        self.pathology_dice.update_state(patho_true, patho_pred)
        self.refined_dice.update_state(patho_true, refined_patho)
        
        return {
            "loss": self.loss_tracker.result(),
            "layer_loss": self.layer_loss_tracker.result(),
            "pathology_loss": self.pathology_loss_tracker.result(),
            "refined_loss": self.refined_loss_tracker.result(),
            "layer_accuracy": self.layer_accuracy.result(),
            "layer_dice": self.layer_dice.result(),
            "pathology_dice": self.pathology_dice.result(),
            "refined_dice": self.refined_dice.result()
        }

    def test_step(self, data):
        x, y = data
        
        layer_true = y[:, :, :, :NUM_LAYERS]
        patho_true = y[:, :, :, NUM_LAYERS:]
        
        layer_pred, patho_pred, refined_patho = self.lapm(x, training=False)
        
        # Compute losses
        layer_loss = tf.reduce_mean(self.focal_loss(layer_true, layer_pred)) + dice_loss(layer_true, layer_pred)
        patho_loss = tf.reduce_mean(self.binary_crossentropy(patho_true, patho_pred)) + dice_loss(patho_true, patho_pred)
        refined_loss = tf.reduce_mean(self.binary_crossentropy(patho_true, refined_patho)) + dice_loss(patho_true, refined_patho)
        total_loss = layer_loss + 2.0 * patho_loss + 1.5 * refined_loss
        
        # Update metrics
        self.loss_tracker.update_state(total_loss)
        self.layer_loss_tracker.update_state(layer_loss)
        self.pathology_loss_tracker.update_state(patho_loss)
        self.refined_loss_tracker.update_state(refined_loss)
        
        self.layer_accuracy.update_state(pixel_accuracy(layer_true, layer_pred))
        self.layer_dice.update_state(layer_true, layer_pred)
        self.pathology_dice.update_state(patho_true, patho_pred)
        self.refined_dice.update_state(patho_true, refined_patho)
        
        return {
            "loss": self.loss_tracker.result(),
            "layer_loss": self.layer_loss_tracker.result(),
            "pathology_loss": self.pathology_loss_tracker.result(),
            "refined_loss": self.refined_loss_tracker.result(),
            "layer_accuracy": self.layer_accuracy.result(),
            "layer_dice": self.layer_dice.result(),
            "pathology_dice": self.pathology_dice.result(),
            "refined_dice": self.refined_dice.result()
        }

# ============= Data Pipeline =============
def create_tf_dataset(dataset_loader, split='train', batch_size=BATCH_SIZE):
    dataset = tf.data.Dataset.from_generator(
        lambda: dataset_loader.data_generator(split),
        output_signature=(
            tf.TensorSpec(shape=(IMG_HEIGHT, IMG_WIDTH, 1), dtype=tf.float32),  # Grayscale input
            tf.TensorSpec(shape=(IMG_HEIGHT, IMG_WIDTH, NUM_LAYERS + NUM_PATHOLOGIES), dtype=tf.float32)
        )
    )
    
    if split == 'train':
        dataset = dataset.shuffle(buffer_size=50)
        # Add augmentation
        dataset = dataset.map(augment_data, num_parallel_calls=tf.data.AUTOTUNE)
    
    dataset = dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return dataset

def augment_data(image, mask):
    """Data augmentation for grayscale OCT images"""
    # Random horizontal flip
    if tf.random.uniform([]) > 0.5:
        image = tf.image.flip_left_right(image)
        mask = tf.image.flip_left_right(mask)
    
    # Random brightness and contrast for grayscale
    image = tf.image.random_brightness(image, 0.1)
    image = tf.image.random_contrast(image, 0.9, 1.1)
    
    # Random rotation (small angles)
    angle = tf.random.uniform([], -0.1, 0.1)  # Â±5.7 degrees
    image = tf.image.rot90(image, k=tf.random.uniform([], 0, 4, dtype=tf.int32))
    mask = tf.image.rot90(mask, k=tf.random.uniform([], 0, 4, dtype=tf.int32))
    
    # Ensure values are in valid range
    image = tf.clip_by_value(image, 0.0, 1.0)
    
    return image, mask

# ============= Enhanced Visualization =============
def create_complete_lapm_visualization(model, image, true_mask=None):
    """Create comprehensive LAPM visualization"""
    # Get predictions
    layer_pred, patho_pred, refined_patho = model.predict(np.expand_dims(image, 0))
    
    layer_map = np.argmax(layer_pred[0], axis=-1)
    pathology_map = np.max(refined_patho[0], axis=-1)
    
    # Create visualization
    fig, axes = plt.subplots(2, 4, figsize=(20, 10))
    
    # Display grayscale image
    display_image = image.squeeze() if image.ndim == 3 else image
    
    # Row 1: Core outputs
    axes[0, 0].imshow(display_image, cmap='gray')
    axes[0, 0].set_title('Original OCT B-Scan', fontsize=14, fontweight='bold')
    axes[0, 0].axis('off')
    
    # Layer segmentation with clinical colors
    layer_colors = plt.cm.Set3(np.linspace(0, 1, NUM_LAYERS))
    colored_layers = layer_colors[layer_map]
    axes[0, 1].imshow(colored_layers)
    axes[0, 1].set_title('ğŸ”¬ Layer Segmentation\n(13 Retinal Layers)', fontsize=14, fontweight='bold')
    axes[0, 1].axis('off')
    
    # Pathology detection
    axes[0, 2].imshow(display_image, cmap='gray', alpha=0.7)
    pathology_overlay = axes[0, 2].imshow(pathology_map, cmap='hot', alpha=0.8)
    axes[0, 2].set_title('ğŸ¯ Disease Detection\n(Pathology Mapping)', fontsize=14, fontweight='bold')
    axes[0, 2].axis('off')
    plt.colorbar(pathology_overlay, ax=axes[0, 2], fraction=0.046)
    
    # Combined LAPM output
    axes[0, 3].imshow(colored_layers, alpha=0.6)
    axes[0, 3].imshow(pathology_map, cmap='Reds', alpha=0.8)
    axes[0, 3].set_title('ğŸ¥ LAPM Analysis\nLayers + Pathology', fontsize=14, fontweight='bold')
    axes[0, 3].axis('off')
    
    # Row 2: Clinical analysis
    layer_pathology_stats = analyze_layer_pathology_complete(layer_map, pathology_map)
    
    # Clinical summary
    axes[1, 0].text(0.05, 0.95, "ğŸ“‹ CLINICAL SUMMARY", fontsize=16, fontweight='bold', 
                   transform=axes[1, 0].transAxes, color='darkblue')
    
    y_pos = 0.85
    total_pathology = 0
    significant_findings = 0
    
    for layer_idx, percentage in layer_pathology_stats.items():
        if percentage > 5:
            layer_name = get_layer_name_complete(layer_idx)
            severity = "ğŸ”´ High" if percentage > 20 else "ğŸŸ¡ Moderate" if percentage > 10 else "ğŸŸ¢ Mild"
            axes[1, 0].text(0.05, y_pos, f"{layer_name}: {percentage:.1f}%", 
                           fontsize=10, transform=axes[1, 0].transAxes)
            axes[1, 0].text(0.05, y_pos-0.04, f"   {severity} involvement", 
                           fontsize=9, transform=axes[1, 0].transAxes, style='italic')
            y_pos -= 0.12
            total_pathology += percentage
            significant_findings += 1
    
    if significant_findings == 0:
        axes[1, 0].text(0.05, 0.7, "âœ… No significant pathology detected", 
                       fontsize=12, transform=axes[1, 0].transAxes, color='green')
    else:
        axes[1, 0].text(0.05, y_pos-0.05, f"ğŸ“Š Total: {total_pathology:.1f}% pathology", 
                       fontsize=11, fontweight='bold', transform=axes[1, 0].transAxes, color='purple')
        axes[1, 0].text(0.05, y_pos-0.12, f"ğŸ” {significant_findings} layers affected", 
                       fontsize=10, transform=axes[1, 0].transAxes, color='red')
    
    axes[1, 0].axis('off')
    
    # Confidence and uncertainty maps
    layer_confidence = np.max(layer_pred[0], axis=-1)
    pathology_confidence = np.max(refined_patho[0], axis=-1)
    
    # Layer confidence
    conf_display = axes[1, 1].imshow(layer_confidence, cmap='viridis')
    axes[1, 1].set_title('ğŸ¯ AI Confidence\n(Layer Prediction)', fontsize=14, fontweight='bold')
    axes[1, 1].axis('off')
    plt.colorbar(conf_display, ax=axes[1, 1], fraction=0.046)
    
    # Treatment recommendations
    axes[1, 2].text(0.05, 0.95, "ğŸ’Š TREATMENT GUIDANCE", fontsize=16, fontweight='bold', 
                   transform=axes[1, 2].transAxes, color='darkgreen')
    
    recommendations = generate_treatment_recommendations_complete(layer_pathology_stats, total_pathology)
    
    y_pos = 0.8
    for i, rec in enumerate(recommendations[:5]):
        priority = "ğŸ”¥ Urgent" if i == 0 and total_pathology > 30 else "â° Routine"
        axes[1, 2].text(0.05, y_pos, f"{i+1}. {rec}", fontsize=10,
                       transform=axes[1, 2].transAxes, color='darkgreen')
        if i == 0 and total_pathology > 20:
            axes[1, 2].text(0.05, y_pos-0.06, f"   {priority}", fontsize=8,
                           transform=axes[1, 2].transAxes, color='red' if 'Urgent' in priority else 'blue')
        y_pos -= 0.14
    
    axes[1, 2].axis('off')
    
    # Performance metrics and model status
    axes[1, 3].text(0.05, 0.9, "ğŸ¤– MODEL STATUS", fontsize=16, fontweight='bold',
                   transform=axes[1, 3].transAxes, color='darkblue')
    
    mean_layer_conf = np.mean(layer_confidence)
    mean_patho_conf = np.mean(pathology_confidence)
    
    axes[1, 3].text(0.05, 0.75, f"Layer Confidence: {mean_layer_conf:.1%}", fontsize=11,
                   transform=axes[1, 3].transAxes)
    axes[1, 3].text(0.05, 0.65, f"Pathology Confidence: {mean_patho_conf:.1%}", fontsize=11,
                   transform=axes[1, 3].transAxes)
    
    if mean_layer_conf > 0.8:
        status = "âœ… High Reliability"
        color = 'green'
    elif mean_layer_conf > 0.6:
        status = "âš ï¸ Moderate Reliability"
        color = 'orange'
    else:
        status = "âŒ Low Reliability"
        color = 'red'
    
    axes[1, 3].text(0.05, 0.45, status, fontsize=12, fontweight='bold',
                   transform=axes[1, 3].transAxes, color=color)
    
    axes[1, 3].text(0.05, 0.25, "ğŸ”¬ U-Net Architecture", fontsize=10,
                   transform=axes[1, 3].transAxes, color='blue')
    axes[1, 3].text(0.05, 0.15, f"ğŸ“Š {model.count_params():,} parameters", fontsize=10,
                   transform=axes[1, 3].transAxes, color='purple')
    
    if true_mask is not None:
        true_layer_map = np.argmax(true_mask[:, :, :NUM_LAYERS], axis=-1)
        layer_accuracy_val = np.mean(layer_map == true_layer_map) * 100
        axes[1, 3].text(0.05, 0.05, f"âœ“ Accuracy: {layer_accuracy_val:.1f}%", fontsize=10,
                       transform=axes[1, 3].transAxes, color='green' if layer_accuracy_val > 70 else 'orange')
    
    axes[1, 3].axis('off')
    
    plt.tight_layout()
    plt.suptitle('Layer-Aware Pathology Mapper (LAPM) - Complete Clinical Analysis', 
                 fontsize=20, fontweight='bold', y=0.98)
    plt.show()
    
    return layer_pathology_stats

def analyze_layer_pathology_complete(layer_map, pathology_map, threshold=0.2):
    """Analyze pathology distribution across layers"""
    stats = {}
    
    for layer_idx in range(NUM_LAYERS):
        layer_mask = (layer_map == layer_idx)
        if np.sum(layer_mask) > 100:  # Only analyze layers with sufficient pixels
            pathology_in_layer = pathology_map[layer_mask]
            pathology_percentage = np.sum(pathology_in_layer > threshold) / len(pathology_in_layer) * 100
            if pathology_percentage > 1:  # Only report meaningful percentages
                stats[layer_idx] = pathology_percentage
    
    return stats

def get_layer_name_complete(layer_idx):
    """Convert layer index to clinical name"""
    layer_names = [
        "Background", "ILM", "NFL", "GCL", "IPL", "INL", 
        "OPL", "ONL", "ELM", "IS", "OS", "RPE", "Choroid"
    ]
    return layer_names[min(layer_idx, len(layer_names)-1)]

def generate_treatment_recommendations_complete(layer_stats, total_pathology):
    """Generate evidence-based treatment recommendations"""
    recommendations = []
    
    if total_pathology > 30:
        recommendations.append("Immediate ophthalmologic referral required")
        recommendations.append("Consider anti-VEGF injection therapy")
    elif total_pathology > 15:
        recommendations.append("Schedule follow-up within 2-4 weeks")
        recommendations.append("Monitor visual acuity changes")
    elif total_pathology > 5:
        recommendations.append("Routine monitoring every 3-6 months")
        recommendations.append("Patient education on symptom awareness")
    else:
        recommendations.append("Continue standard care protocol")
        recommendations.append("Annual comprehensive eye examination")
    
    # Layer-specific recommendations
    for layer_idx, percentage in layer_stats.items():
        if percentage > 15:
            layer_name = get_layer_name_complete(layer_idx)
            if layer_idx in [11, 12]:  # RPE, Choroid
                recommendations.append(f"Evaluate {layer_name} for AMD progression")
            elif layer_idx in [8, 9, 10]:  # ELM, IS, OS
                recommendations.append(f"Assess {layer_name} photoreceptor integrity")
    
    return recommendations[:6]

# ============= Main Training Script =============
if __name__ == "__main__":
    try:
        print("ğŸš€ Starting Complete LAPM Training...")
        
        # GPU setup
        try:
            gpus = tf.config.list_physical_devices('GPU')
            if gpus:
                for gpu in gpus:
                    tf.config.experimental.set_memory_growth(gpu, True)
                print("âœ… GPU optimization enabled")
            else:
                print("âš ï¸ No GPU found, using CPU (training will be slower)")
        except Exception as e:
            print(f"âš ï¸ GPU setup failed: {e}")
        
        tf.keras.backend.clear_session()
        
        print("ğŸ“ Loading OCTAVE dataset...")
        dataset_loader = OCTAVEDatasetFromZip(ZIP_PATH)
        
        train_dataset = create_tf_dataset(dataset_loader, split='train')
        val_dataset = create_tf_dataset(dataset_loader, split='val')
        
        print("ğŸ—ï¸ Building Complete LAPM model...")
        lapm_model = build_complete_lapm()
        lapm_model.summary()
        
        print(f"ğŸ“Š Model Parameters: {lapm_model.count_params():,}")
        print(f"ğŸ’¾ Estimated GPU Memory: ~{lapm_model.count_params() * 4 / 1e9:.2f} GB")
        
        print("ğŸ¯ Initializing Complete Trainer...")
        trainer = CompleteLAPMTrainer(lapm_model)
        
        # Enhanced optimizer with learning rate scheduling
        initial_learning_rate = 5e-5  # Conservative learning rate
        lr_schedule = tf.keras.optimizers.schedules.CosineDecayRestarts(
            initial_learning_rate,
            first_decay_steps=1000,
            t_mul=2.0,
            m_mul=0.8,
            alpha=1e-7
        )
        
        trainer.compile(
            optimizer=tf.keras.optimizers.AdamW(
                learning_rate=lr_schedule,
                weight_decay=1e-5
            )
        )
        
        print("ğŸ“ˆ Starting Enhanced Training with Complete Architecture...")
        
        # Advanced callbacks
        callbacks = [
            tf.keras.callbacks.EarlyStopping(
                monitor='val_layer_dice',
                patience=15,
                restore_best_weights=True,
                mode='max',
                verbose=1
            ),
            tf.keras.callbacks.ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.3,
                patience=7,
                min_lr=1e-8,
                verbose=1
            ),
            tf.keras.callbacks.ModelCheckpoint(
                'complete_lapm_best.h5',
                monitor='val_layer_dice',
                save_best_only=True,
                mode='max',
                verbose=1
            ),
            tf.keras.callbacks.TensorBoard(
                log_dir='./logs',
                histogram_freq=1,
                write_graph=True,
                update_freq='epoch'
            )
        ]
        
        # Train model with progress tracking
        print("â° Training Progress:")
        print("Expected time: ~2-4 hours (CPU) or ~30-60 minutes (GPU)")
        
        history = trainer.fit(
            train_dataset,
            validation_data=val_dataset,
            epochs=EPOCHS,
            callbacks=callbacks,
            verbose=1
        )
        
        print("ğŸ‰ Training completed successfully!")
        
        # Demonstrate comprehensive visualization
        print("ğŸ” Generating Complete LAPM Visualization...")
        for images, masks in val_dataset.take(1):
            sample_image = images[0].numpy()
            sample_mask = masks[0].numpy()
            
            print("ğŸ¨ Creating Clinical Analysis Dashboard...")
            stats = create_complete_lapm_visualization(lapm_model, sample_image, sample_mask)
            
            print("\nğŸ“‹ Clinical Summary:")
            total_pathology = sum(stats.values())
            if total_pathology > 20:
                print(f"  ğŸ”´ HIGH CONCERN: {total_pathology:.1f}% total pathology detected")
                print("  ğŸ“ Recommend immediate specialist consultation")
            elif total_pathology > 10:
                print(f"  ğŸŸ¡ MODERATE: {total_pathology:.1f}% pathology detected")
                print("  ğŸ“… Schedule follow-up within 2-4 weeks")
            elif total_pathology > 2:
                print(f"  ğŸŸ¢ MILD: {total_pathology:.1f}% pathology detected")  
                print("  ğŸ‘ï¸ Continue routine monitoring")
            else:
                print("  âœ… NO SIGNIFICANT PATHOLOGY: Normal retinal structure")
                print("  ğŸ“‹ Annual examination recommended")
            
            print("\nğŸ”¬ Layer-by-Layer Analysis:")
            for layer, percentage in stats.items():
                layer_name = get_layer_name_complete(layer)
                print(f"    {layer_name}: {percentage:.1f}% affected")
            
            break
        
        # Save model and training history
        lapm_model.save('complete_lapm_final.h5')
        
        # Save training history
        import pickle
        with open('training_history.pkl', 'wb') as f:
            pickle.dump(history.history, f)
        
        print("ğŸ’¾ Complete LAPM model and training history saved!")
        
        # Training performance summary
        final_layer_dice = history.history['val_layer_dice'][-1]
        final_pathology_dice = history.history['val_pathology_dice'][-1]
        final_accuracy = history.history['val_layer_accuracy'][-1]
        
        print("\nğŸ“Š Final Performance Summary:")
        print(f"  Layer Segmentation Dice: {final_layer_dice:.3f}")
        print(f"  Pathology Detection Dice: {final_pathology_dice:.3f}")
        print(f"  Overall Accuracy: {final_accuracy:.1%}")
        
        if final_layer_dice > 0.7:
            print("  âœ… EXCELLENT: Clinical-grade performance achieved!")
        elif final_layer_dice > 0.5:
            print("  âš¡ GOOD: Suitable for clinical assistance")
        else:
            print("  âš ï¸ DEVELOPING: Continue training for better performance")
        
        # Plot comprehensive training history
        plt.figure(figsize=(20, 12))
        
        # Loss plots
        plt.subplot(3, 4, 1)
        plt.plot(history.history['loss'], label='Training Loss', linewidth=2)
        plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
        plt.title('Overall Training Loss', fontsize=14, fontweight='bold')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        plt.subplot(3, 4, 2)
        plt.plot(history.history['layer_loss'], label='Layer Loss', linewidth=2, color='blue')
        plt.plot(history.history['val_layer_loss'], label='Val Layer Loss', linewidth=2, color='lightblue')
        plt.title('Layer Segmentation Loss', fontsize=14, fontweight='bold')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        plt.subplot(3, 4, 3)
        plt.plot(history.history['pathology_loss'], label='Pathology Loss', linewidth=2, color='red')
        plt.plot(history.history['val_pathology_loss'], label='Val Pathology Loss', linewidth=2, color='lightcoral')
        plt.title('Pathology Detection Loss', fontsize=14, fontweight='bold')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        plt.subplot(3, 4, 4)
        plt.plot(history.history['refined_loss'], label='Refined Loss', linewidth=2, color='green')
        plt.plot(history.history['val_refined_loss'], label='Val Refined Loss', linewidth=2, color='lightgreen')
        plt.title('Refined Pathology Loss', fontsize=14, fontweight='bold')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # Accuracy plots
        plt.subplot(3, 4, 5)
        plt.plot(history.history['layer_accuracy'], label='Layer Accuracy', linewidth=2, color='purple')
        plt.plot(history.history['val_layer_accuracy'], label='Val Layer Accuracy', linewidth=2, color='plum')
        plt.title('Layer Segmentation Accuracy', fontsize=14, fontweight='bold')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # Dice coefficient plots
        plt.subplot(3, 4, 6)
        plt.plot(history.history['layer_dice'], label='Layer Dice', linewidth=2, color='navy')
        plt.plot(history.history['val_layer_dice'], label='Val Layer Dice', linewidth=2, color='skyblue')
        plt.title('Layer Segmentation Dice', fontsize=14, fontweight='bold')
        plt.xlabel('Epoch')
        plt.ylabel('Dice Score')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        plt.subplot(3, 4, 7)
        plt.plot(history.history['pathology_dice'], label='Pathology Dice', linewidth=2, color='darkred')
        plt.plot(history.history['val_pathology_dice'], label='Val Pathology Dice', linewidth=2, color='salmon')
        plt.title('Pathology Detection Dice', fontsize=14, fontweight='bold')
        plt.xlabel('Epoch')
        plt.ylabel('Dice Score')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        plt.subplot(3, 4, 8)
        plt.plot(history.history['refined_dice'], label='Refined Dice', linewidth=2, color='darkgreen')
        plt.plot(history.history['val_refined_dice'], label='Val Refined Dice', linewidth=2, color='lightgreen')
        plt.title('Refined Pathology Dice', fontsize=14, fontweight='bold')
        plt.xlabel('Epoch')
        plt.ylabel('Dice Score')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # Learning rate
        if 'lr' in history.history or 'learning_rate' in history.history:
            lr_key = 'lr' if 'lr' in history.history else 'learning_rate'
            plt.subplot(3, 4, 9)
            plt.plot(history.history[lr_key], linewidth=2, color='orange')
            plt.title('Learning Rate Schedule', fontsize=14, fontweight='bold')
            plt.xlabel('Epoch')
            plt.ylabel('Learning Rate')
            plt.yscale('log')
            plt.grid(True, alpha=0.3)
        
        # Performance summary
        plt.subplot(3, 4, 10)
        performance_metrics = [
            final_layer_dice,
            final_pathology_dice, 
            final_accuracy
        ]
        metric_names = ['Layer Dice', 'Pathology Dice', 'Accuracy']
        colors = ['blue', 'red', 'green']
        
        bars = plt.bar(metric_names, performance_metrics, color=colors, alpha=0.7)
        plt.title('Final Performance Metrics', fontsize=14, fontweight='bold')
        plt.ylabel('Score')
        plt.ylim(0, 1)
        
        # Add value labels on bars
        for bar, value in zip(bars, performance_metrics):
            plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
                    f'{value:.3f}', ha='center', va='bottom', fontweight='bold')
        
        plt.grid(True, alpha=0.3)
        plt.xticks(rotation=45)
        
        # Model architecture summary
        plt.subplot(3, 4, 11)
        plt.text(0.1, 0.8, 'LAPM Architecture Summary', fontsize=14, fontweight='bold')
        plt.text(0.1, 0.6, f'Total Parameters: {lapm_model.count_params():,}', fontsize=10)
        plt.text(0.1, 0.5, f'Input Size: {IMG_HEIGHT}Ã—{IMG_WIDTH}Ã—1', fontsize=10)
        plt.text(0.1, 0.4, f'Output Layers: {NUM_LAYERS}', fontsize=10)
        plt.text(0.1, 0.3, f'Pathology Classes: {NUM_PATHOLOGIES}', fontsize=10)
        plt.text(0.1, 0.2, 'Architecture: U-Net with Attention', fontsize=10)
        plt.text(0.1, 0.1, 'Loss: Focal + Dice Combined', fontsize=10)
        plt.axis('off')
        
        # Training summary
        plt.subplot(3, 4, 12)
        plt.text(0.1, 0.8, 'Training Summary', fontsize=14, fontweight='bold')
        plt.text(0.1, 0.6, f'Epochs: {len(history.history["loss"])}', fontsize=10)
        plt.text(0.1, 0.5, f'Batch Size: {BATCH_SIZE}', fontsize=10)
        plt.text(0.1, 0.4, f'Optimizer: AdamW', fontsize=10)
        plt.text(0.1, 0.3, f'Initial LR: {initial_learning_rate}', fontsize=10)
        
        if final_layer_dice > 0.7:
            status_text = 'âœ… Clinical Grade'
            status_color = 'green'
        elif final_layer_dice > 0.5:
            status_text = 'âš¡ Clinical Ready'
            status_color = 'orange'
        else:
            status_text = 'âš ï¸ In Development'
            status_color = 'red'
            
        plt.text(0.1, 0.1, f'Status: {status_text}', fontsize=10, color=status_color, fontweight='bold')
        plt.axis('off')
        
        plt.suptitle('Complete Layer-Aware Pathology Mapper (LAPM) Training Results', 
                     fontsize=18, fontweight='bold')
        plt.tight_layout()
        plt.show()
        
        print("\nğŸ† LAPM Training Complete!")
        print("ğŸ”¬ Your Layer-Aware Pathology Mapper is ready for clinical OCT analysis!")
        print("ğŸ“Š Use the saved model for real-time pathology detection and layer segmentation!")
        
    except Exception as e:
        print(f"âŒ Training failed with error: {e}")
        import traceback
        traceback.print_exc()
```
